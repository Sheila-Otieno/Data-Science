```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import datetime as dt
```

```python
air_1987 = pd.read_csv('1987.csv')
air_1987.head()
```

```python
air_1987_data = air_1987[['Year','Month','DayofMonth','DayOfWeek','DepTime','CRSDepTime','ArrTime','CRSArrTime','UniqueCarrier',
                         'FlightNum','ActualElapsedTime','CRSElapsedTime','ArrDelay','DepDelay','Origin','Dest','Distance',
                         'Cancelled','Diverted']]
```

```python
air_1987_data.head()
```

```python
air_1987_data.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
#correlation matrix
air_1987_data_corr = air_1987_data.corr()
#column names
col = ['Year','Month','DayofMonth','DayOfWeek','DepTime','CRSDepTime','ArrTime','CRSArrTime','UniqueCarrier',
                         'FlightNum','ActualElapsedTime','CRSElapsedTime','ArrDelay','DepDelay','Origin','Dest','Distance',
                         'Cancelled','Diverted']  
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(air_1987_data_corr, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(col)
ax.set_yticklabels(col)
plt.show()
```

Read the 1997 dataset using pandas library

```python
air_1997 = pd.read_csv('1997.csv')
air_1997.head()
```

```python
air_1997.columns
```

```python
air_1997_data = air_1997[['Year','Month','DayofMonth','DayOfWeek','DepTime','CRSDepTime','ArrTime','CRSArrTime','UniqueCarrier',
                         'FlightNum','ActualElapsedTime','CRSElapsedTime','ArrDelay','DepDelay','Origin','Dest','Distance',
                         'Cancelled','Diverted']]
air_1997_data.head()
```

```python
#combining the year, month and dayofmonth column to create a date column 
air_1997_data['Date'] = air_1997_data['Year'].map(str) +'-'+ air_1997_data['Month'].map(str) +'-'+ air_1997_data['DayofMonth'].map(str)
air_1997_data['Date'].head(10)
air_1997_data.head()
```

```python
air_1997_data.dtypes
```

```python
air_1997_data['CRSArrTime'] = pd.to_datetime(air_1997_data['CRSArrTime'])
air_1997_data.dtypes
```

```python
air_1997_data['Date']= pd.to_datetime(air_1997_data['Date'])
air_1997_data.dtypes
```

```python
air_1997_data['DepTime'] = air_1997_data['DepTime'].astype(str)
air_1997_data.dtypes
```

```python
air_1997_data['DepTime'].head()
```

```python
air_1997_data['DepTime'] = air_1997_data['DepTime'].apply(lambda x: x.strip('.0'))
air_1997_data['DepTime'].head()
```

```python
air_1997_data['DepTime'] = air_1997_data['DepTime'].apply(lambda x: f'{x[:-2]}:{x[-2:]}')
air_1997_data['DepTime'].head()
```

```python
air_1997_data['DepTime'].dropna()
```

```python
air_1997_data.head()
```

```python
air_1997_data.isnull().sum()
```

```python
air_1997_data['DepTime'] = air_1997_data['DepTime'].apply(lambda x: x.replace('n:an','0:00'))
```

```python
air_1997_data['DepTime'] = air_1997_data['DepTime'].apply(lambda x: x.strip(':7'))
```

```python
air_1997_data['DepTime'] = air_1997_data['DepTime'].apply(lambda x: x.strip(':75'))
```
